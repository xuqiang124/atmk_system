```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 读取CSV文件
df = pd.read_csv('logs/DA-20k/cut/LHABS/voyiyu-labs-0604-122607/training.csv')
# 筛选出所需的列
epochs_labs = df['epoch'].tolist()
train_loss_labs = df['lcm_accuracy_3k'].tolist()
val_loss_labs = df['val_lcm_accuracy_3k'].tolist()
# max_val_loss_labs = max(max(train_loss_labs),max(val_loss_labs))

# lbs
df = pd.read_csv('logs/DA-20k/cut/LBS/dgfzzb-lbs-0604-124454/training.csv')
epochs_lbs = df['epoch'].tolist()
train_loss_lbs = df['lcm_accuracy_3k'].tolist()
val_loss_lbs = df['val_lcm_accuracy_3k'].tolist()

# lhab
df = pd.read_csv('logs/DA-20k/cut/LHAB/ynukky-lab-0604-133426/training.csv')
epochs_lab = df['epoch'].tolist()
train_loss_lab = df['accuracy_3k'].tolist()
val_loss_lab = df['val_accuracy_3k'].tolist()

# Basic
df = pd.read_csv('logs/DA-20k/cut/Basic/gkrqzx-b-0608-122552/training.csv')
epochs_basic = df['epoch'].tolist()
train_loss_basic = df['accuracy_3k'].tolist()
val_loss_basic = df['val_accuracy_3k'].tolist()

# 确保横轴最大值显示为120
max_epochs = 120

# 创建图形
fig, axs = plt.subplots(2, 2, figsize=(5, 5))

# 绘制LHABS子图
axs[0, 0].plot(epochs_labs, train_loss_labs, 'b-', label='Train',linewidth=3)
axs[0, 0].plot(epochs_labs, val_loss_labs, 'orange', label='Val',linewidth=3)
# axs[0, 0].axhline(y=0.196396186947822, xmin=0, xmax=120, color='red', linestyle='--', linewidth=1, label='Test')
axs[0, 0].set_title('LHABS')
axs[0, 0].set_xlabel('Epoch')
axs[0, 0].set_ylabel('Accuracy')
axs[0, 0].set_xlim(0, max_epochs)
axs[0, 0].set_yticks(np.linspace(0,1,3))
axs[0, 0].set_xticks(np.linspace(0,120,4))
axs[0, 0].legend()

# 绘制LHAB子图
axs[0, 1].plot(epochs_lab, train_loss_lab, 'b-', label='Train',linewidth=3)
axs[0, 1].plot(epochs_lab, val_loss_lab, 'orange', label='Val',linewidth=3)
axs[0, 1].set_title('LHAB')
axs[0, 1].set_xlabel('Epoch')
axs[0, 1].set_ylabel('Accuracy')
axs[0, 1].set_xlim(0, max_epochs)
axs[0, 1].set_yticks(np.linspace(0,1,3))
axs[0, 1].set_xticks(np.linspace(0,120,4))
axs[0, 1].legend()

# 绘制LBS子图
axs[1, 0].plot(epochs_lbs, train_loss_lbs, 'b-', label='Train',linewidth=3)
axs[1, 0].plot(epochs_lbs, val_loss_lbs, 'orange', label='Val',linewidth=3)
axs[1, 0].set_title('LBS')
axs[1, 0].set_xlabel('Epoch')
axs[1, 0].set_ylabel('Accuracy')
axs[1, 0].set_xlim(0, max_epochs)
axs[1, 0].set_yticks(np.linspace(0,1,3))
axs[1, 0].set_xticks(np.linspace(0,120,4))
axs[1, 0].legend()

# 绘制Basic子图
axs[1, 1].plot(epochs_basic, train_loss_basic, 'b-', label='Train',linewidth=3)
axs[1, 1].plot(epochs_basic, val_loss_basic, 'orange', label='Val',linewidth=3)
axs[1, 1].set_title('Basic')
axs[1, 1].set_xlabel('Epoch')
axs[1, 1].set_ylabel('Accuracy')
axs[1, 1].set_xlim(0, max_epochs)
axs[1, 1].set_yticks(np.linspace(0,1,3))
axs[1, 1].set_xticks(np.linspace(0,120,4))
axs[1, 1].legend()

# 调整子图间距
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 读取CSV文件
df = pd.read_csv('logs/DA-20k/cut/LHABS/voyiyu-labs-0604-122607/training.csv')
# 筛选出所需的列
epochs_labs = df['epoch'].tolist()
train_loss_labs = df['lcm_accuracy_3k'].tolist()
val_loss_labs = df['val_lcm_accuracy_3k'].tolist()
# max_val_loss_labs = max(max(train_loss_labs),max(val_loss_labs))

# lbs
df = pd.read_csv('logs/DA-20k/cut/LBS/dgfzzb-lbs-0604-124454/training.csv')
epochs_lbs = df['epoch'].tolist()
train_loss_lbs = df['lcm_accuracy_3k'].tolist()
val_loss_lbs = df['val_lcm_accuracy_3k'].tolist()

# lab
df = pd.read_csv('logs/DA-20k/cut/LHAB/ynukky-lab-0604-133426/training.csv')
epochs_lab = df['epoch'].tolist()
train_loss_lab = df['accuracy_3k'].tolist()
val_loss_lab = df['val_accuracy_3k'].tolist()

# Basic
df = pd.read_csv('logs/DA-20k/cut/Basic/gkrqzx-b-0608-122552/training.csv')
epochs_basic = df['epoch'].tolist()
train_loss_basic = df['accuracy_3k'].tolist()
val_loss_basic = df['val_accuracy_3k'].tolist()

# 确保横轴最大值显示为120
max_epochs = 120
font1 = {'family' : 'Times New Roman', 'weight' : 'normal', 'size' : 10}
# 创建图形
fig, axs = plt.subplots(1, 4, figsize=(17/2.54, 4.5/2.54), dpi=300)

for i in range(4):
    axs[i].tick_params(axis='both', labelsize=8, labelrotation=0, labelcolor='black', which='both', direction='in', pad=5, labelbottom=True, labelleft=True, bottom=True, left=True, right=False, top=False, labeltop=False, labelright=False)
    for label in (axs[i].get_xticklabels() + axs[i].get_yticklabels()):
        label.set_fontname('Times New Roman')

# 绘制LHABS子图
axs[0].plot(epochs_labs, train_loss_labs, 'b-', label='Train',linewidth=2)
axs[0].plot(epochs_labs, val_loss_labs, 'orange', label='Val',linewidth=2)
# axs[0].axhline(y=0.196396186947822, xmin=0, xmax=120, color='red', linestyle='--', linewidth=1, label='Test')
axs[0].set_title('LHABS',font1)
axs[0].set_xlabel('Epoch',font1)
axs[0].set_ylabel('Accuracy',font1)
axs[0].set_xlim(0, max_epochs)
axs[0].set_yticks(np.linspace(0,1,3))
axs[0].set_xticks(np.linspace(0,120,4))
axs[0].legend(loc='upper right', prop={'family': 'Times New Roman', 'size': 6})

# 绘制LHAB子图
axs[1].plot(epochs_lab, train_loss_lab, 'b-', label='Train',linewidth=2)
axs[1].plot(epochs_lab, val_loss_lab, 'orange', label='Val',linewidth=2)
axs[1].set_title('LHAB',font1)
axs[1].set_xlabel('Epoch',font1)
axs[1].set_ylabel('Accuracy',font1)
axs[1].set_xlim(0, max_epochs)
axs[1].set_yticks(np.linspace(0,1,3))
axs[1].set_xticks(np.linspace(0,120,4))
axs[1].legend(loc='upper right', prop={'family': 'Times New Roman', 'size': 6})

# 绘制LBS子图
axs[2].plot(epochs_lbs, train_loss_lbs, 'b-', label='Train',linewidth=2)
axs[2].plot(epochs_lbs, val_loss_lbs, 'orange', label='Val',linewidth=2)
axs[2].set_title('LBS',font1)
axs[2].set_xlabel('Epoch',font1)
axs[2].set_ylabel('Accuracy',font1)
axs[2].set_xlim(0, max_epochs)
axs[2].set_yticks(np.linspace(0,1,3))
axs[2].set_xticks(np.linspace(0,120,4))
axs[2].legend(loc='upper right', prop={'family': 'Times New Roman', 'size': 6})

# 绘制Basic子图
axs[3].plot(epochs_basic, train_loss_basic, 'b-', label='Train',linewidth=2)
axs[3].plot(epochs_basic, val_loss_basic, 'orange', label='Val',linewidth=2)
axs[3].set_title('Basic',font1)
axs[3].set_xlabel('Epoch',font1)
axs[3].set_ylabel('Accuracy',font1)
axs[3].set_xlim(0, max_epochs)
axs[3].set_yticks(np.linspace(0,1,3))
axs[3].set_xticks(np.linspace(0,120,4))
axs[3].legend(loc='upper left', prop={'family': 'Times New Roman', 'size': 6})

# 调整子图间距
plt.tight_layout()

plt.savefig('convergence_speed.pdf',dpi=300,bbox_inches = 'tight')
plt.show()
```
